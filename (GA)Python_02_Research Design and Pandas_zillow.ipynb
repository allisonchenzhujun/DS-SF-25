```python
#Scale all $value to $. 
df.SalePriceUnit.unique()
```

```python
#Scale all price into the same unit. 
df_1 = df[df.SalePriceUnit == '$']
df_2 = df[df.SalePriceUnit == '$M']
#here the axis =1 apply a method across each row. = 0, apply to each column. 
df_1 = df_1.drop('SalePriceUnit',axis = 1)
df_2 = df_2.drop('SalePriceUnit',axis = 1)

#Replace the content in a column
df_2.SalePrice = df_2.SalePrice * (10 **6)
df = pd.concat([df_1,df_2])
```

```python
df.sort_index()
```

```python
df.drop(['Latitude','Longitude'],axis =1, inplace = True)
```

```python
# Convert Boolean to binary variable
df.IsAStudio = 1*df['IsAStudio']
```

```python
#Size:
df.SizeUnit.unique()
#not a useful varible, sqft or Nan, can simply just drop this. 
df.drop('SizeUnit', axis = 1, inplace = True)
```

```python
#Scale LotSize
df.LotSizeUnit.unique()
#convert acre to sqft.
```

```python
#Group1: the NA values 
df_na = df[df.LotSizeUnit.isnull()]
df_na
```

```python
#Group 2: The Sqft values 
df_sqft = df[df.LotSizeUnit == 'sqft']
df_sqft = df_sqft.drop('LotSizeUnit', axis =1)
df_sqft.shape
```

```python
#Group 3: The Ac values
df_ac = df[df.LotSizeUnit == 'ac']
df_ac = df_ac.drop('LotSizeUnit', axis = 1)
df_ac.shape
```

```python
#Scale ac to square
# 1 acre = 43,560 sqft
df_ac.LotSize = df_ac.LotSize * 43560
```

```python
df = pd.concat([df_na,df_sqft,df_ac]).sort_index()
df = df.drop(['LotSizeUnit'], axis =1)
df
```
